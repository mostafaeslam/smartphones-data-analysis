# Introduction 

 • This dataset provides information on various smartphone models including their specifications and expert ratings.</br> 
 • It serves as a valuable resource for analyzing the correlation between specifications and overall ratings of smartphones.

 I'm keenly interested in working on this smartphone dataset due to my fascination with technology and innovation in the mobile industry.

The dataset can be found on Kaggle : https://www.kaggle.com/datasets/mousamkumar123/smartphone-uncleaned-dataset/ </br>

The dataset comprises 1020 records and encompasses 9 attributes.

The Attributes are:
- **Unnamed :** Index
- **Model Name :** The name or model number of the smartphone. </br>
- **Price :** The retail price of the smartphone at the time of data collection. </br>
- **Spec Score:** A numerical score representing the overall specifications and features of the smartphone. </br>
- **Processor:** Details about the processor (CPU) used in the smartphone. </br>
- **Storage:** The storage capacity of the smartphone in gigabytes. </br>
- **Battery:** The capacity of the battery in milliampere-hours (mAh). </br>
- **Display Features:** Information about the display features such as size, resolution, and type. </br>
- **Camera:** Details about the camera specifications including megapixels, aperture, and additional features.

# Data cleaning

```python
# drop Unnamed column because it unnecessary
Sp_df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
Sp_df.sample(5)
```

```python
# rename the columns in the dataset 
Sp_df.rename(columns={'model name': 'Model Name', 'price': 'Price', 
                   'spec score': 'specification score', 'processor': 'Processor', 
                   'storage': 'Storage', 'battery': 'Battery', 
                   'display features': 'Display Features' , 'camera': 'Camera '}, inplace=True)
```

```python
Sp_df.sample(5)
```

```python
# check for duplicate values in the dataset
duplicate_values = Sp_df.duplicated().sum()
print("Number of duplicate values: ", duplicate_values)
```

```python
# remove duplicate values from the dataset
Sp_df.drop_duplicates(inplace=True) 
```

```python
# check for duplicate values in the dataset
duplicate_values = Sp_df.duplicated().sum()
print("Number of duplicate values: ", duplicate_values)
```

```python
# check for missing values in the dataset
missing_values = Sp_df.isnull().sum()
print(missing_values)
```

```python
# drop the missing values in camera row and specification score	
Sp_df.dropna(subset=['Camera ', 'specification score'], inplace=True)
```

```python
# check for missing values in the dataset 
missing_values = Sp_df.isnull().sum()
print(missing_values)
```

```python
# change price and specefications score from string into numeric values
Sp_df['Price'] = Sp_df['Price'].str.replace('₹', '') 
Sp_df['Price'] = Sp_df['Price'].str.replace(',', '')
Sp_df['Price'] = pd.to_numeric(Sp_df['Price'])
Sp_df['specification score'] = pd.to_numeric(Sp_df['specification score'])
```

```python
Sp_df.sample(5)
```

```python
# Splitting the "Main Cast" column into "Director" and "Actors & Actresses" attributes columns
Sp_df[["RAM", "Storage"]] = Sp_df["Storage"].str.split(
    ",", expand=True
)
```

```python
Sp_df.sample(5)
```

```python
# Removing the "RAM" from the "RAM"column
Sp_df["RAM"] = Sp_df["RAM"].str.replace("RAM", "")
```

```python
# Reordering the columns
Sp_df = Sp_df[["Model Name",
                "Price",
                  "specification score",
                    "Processor",
                      "RAM",
                        "Storage",
                          "Battery",
                            "Display Features",
                              "Camera "]]
```

## Data Visualization

```python
# Plotting the correlation matrix of the numerical values in the dataset
import seaborn as sns

# Select only numeric columns for correlation calculation
numeric_data = Sp_df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
corr_matrix = numeric_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plotting the histogram of the Price attribute

# Plot series of the price of the phones over the specification score
Sp_df.groupby("specification score")["Price"].mean().plot(figsize=(20, 6), color="green")

# Setting the plot title
plt.title("\nAverage Mobile price per specification score\n", weight="bold")

# Setting the x-axis label
plt.xlabel("\nScore")

# Setting the y-axis label
plt.ylabel("Price\n")

plt.legend()
plt.show()
```

```python
# Plotting the histogram of the "Processor" attribute

# Get the top 10 used Processors
top_Processor = Sp_df["Processor"].value_counts().nlargest(10)

# Replace the other Processors with "others"
processor_counts = (
    Sp_df["Processor"].apply(lambda x: x if x in top_Processor else "Other Processors").value_counts()
)

# Plotting the pie chart
processor_counts.plot.pie(
    figsize=(12, 10),
    autopct="%.2f%%",
    fontsize=12,
    textprops={"color": "blue"},
)

# Setting the plot title
plt.title("\nNumber of phones per Processor\n", weight="bold")

# Adding genre names to the legend
plt.legend(labels=processor_counts.index)

plt.show()
```

```python
# remove the "GB" from the "RAM" column
Sp_df["RAM"] = Sp_df["RAM"].replace('[\D]', '', regex=True).astype(float)
Sp_df["RAM"] = pd.to_numeric(Sp_df["RAM"])

# Plot series of the specification score of the phones over the RAM
Sp_df.groupby("RAM")["specification score"].mean().plot(figsize=(20, 6), color="purple")

# Setting the plot title
plt.title("\nAverage specification score per RAM\n", weight="bold")

# Setting the x-axis label
plt.xlabel("\nRAM")

# Setting the y-axis label
plt.ylabel("specification score\n")

plt.legend()
plt.show()
```

```python
# Plotting the histogram of the RAM attribute

# Plot series of the Price of the phones over the RAM
Sp_df.groupby("RAM")["Price"].mean().plot(figsize=(20, 6), color="purple")

# Setting the plot title
plt.title("\nAverage Price per RAM\n", weight="bold")

# Setting the x-axis label
plt.xlabel("\nRAM")

# Setting the y-axis label
plt.ylabel("Price\n")

plt.legend()
plt.show()
```
